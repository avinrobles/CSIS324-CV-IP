##Working Images - Grayscaling

Grayscaling
  - is the process of converting an image from other color spaces
    - RGB
      - red, green, blue
      -- is an additive color model used to represent colors on electronic displays suc ha s monitors,
      television screen, and digital camera
        -- values ranging from 0 to 255
    - CMYK
      - Cyan, Magenta, Yellow, Key/Black
      -- is a subtractive color model used in color printing and reproduction process synch as
      offset printing, digital printing, and color photography
    - HSV
      -- Hue, Saturation, Value
      ---> Parameters
      - this is a cylindrical color model that represents color based on these given parameters:
        - Hue
          -- Refers to the dominant wavelength of light and represent the perceived [red, gree, and blue]
        - Saturation
          -- Refers to intensity or purity of the color. The higher intensity = more vivid or intense color
        - Value
          -- represents the brightness or lightness of color. The higher the value == more bright while lower the value == more darker
      

##Importance of grayscaling
  - Dimension reduction
    - grayscale images are single-dimension
  - Reduces model complexity
  - For other algo works

```python
import cv2
import numpy as np
from skimage.color import rgb2lab, lab2rgb
from skimage.io import imread
import matplotlib.pylab as plt
from google.colab.patches import cv2_imshow
from google.colab import files
```

```python
a = np.array([1, 2, 5, 9, 11])
a = a.reshape(1, -1)

plt.imshow(a, cmap = 'gray')
```

```python
#reading values from 0 to 255
values = np.arange(0,255)

#converting 1D values to 2D
values = values.reshape(1, -1)
plt.imshow(values, cmap = 'gray')
```

```python
#reading values from 0 to 255
values = np.arange(0,255)

#converting 1D values to 2D
values = values.reshape(1, -1)
plt.imshow(values, cmap = 'gray')

repeat = np.repeat(values, repeats = 100, axis = 0)
plt.imshow(repeat)
```

```python
def read_file(filename):
  image = cv2.imread(filename)
  if image is None:
    raise ValueError('Unable to read the image. Check the file path or format.')
  cv2_imshow(image)
  return image

#upload the image file
uploaded = files.upload()

#get the filename of the uploaded image
filename = next(iter(uploaded))

#read and display the uploaded file
image = read_file(filename)
```

Color images using OpenCV
  - rgb color space
    - are commonly found describing the physical output of display such as televiosn screens or monitors

```python
image = rgb2lab(image)
```

```python
image[..., 1] = image[...,2] = 0
```

```python
image = lab2rgb(image)
```

```python
plt.figure(figsize = (20, 10))
plt.subplot(122), plt.imshow(image), plt.axis('off')
plt.title("Grayscale Image")
plt.show()
```

```python
plt.subplot(2,2,1)
plt.title("R Channel")
plt.imshow(image[: , : , 0], cmap = 'gray')

plt.subplot(2,2,2)
plt.title("G Channel")
plt.imshow(image[: , : , 1], cmap = 'gray')

plt.subplot(2,2,3)
plt.title("B Channel")
plt.imshow(image[: , : , 2], cmap = 'gray')

plt.subplot(2,2,4)
plt.title("Original Channel")
plt.imshow(image)
```
